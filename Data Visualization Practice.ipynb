```python
# PREDIÇÃO DO PREÇO DE UM CARRO COM HIGHWAY-MPG = 30

linreg_model3 = lr()
linreg_model3.fit(df[['highway-mpg']], df['price'])
predicao = linreg_model3.predict(np.array(30.0).reshape(-1,1))

intercept2 = linreg_model3.intercept_
coef2 = linreg_model3.coef_ * -1 # multiplicando por -1 apenas para mostrar na fórmula abaixo a subtração
price30mpg = intercept2 - coef2 * 30.0

print('USD: ', price30mpg)
print('USD: ', predicao)
```
